```python
import pandas as pd

# Load the new data
data_path = 'C:/Users/jemal/Downloads/archive1/WA_Fn-UseC_-Telco-Customer-Churn.csv'
df = pd.read_csv(data_path)

# Select a subset of the data for testing
# Randomly sample 10 rows and exclude the Churn column
subset_random = df.sample(n=10, random_state=42).drop(columns=["Churn"])

# Alternatively, filter specific conditions (e.g., "Month-to-month" contract) and exclude the Churn column
subset_filtered = df[df["Contract"] == "Month-to-month"].sample(n=10, random_state=42).drop(columns=["Churn"])

# Save the subset to a new file for testing
subset_random.to_csv("subset_random_test.csv", index=False)
subset_filtered.to_csv("subset_filtered_test.csv", index=False)

# Display the subsets
print("Random Sample Subset (Without Churn Column):")
print(subset_random)

print("\nFiltered Subset (Month-to-month, Without Churn Column):")
print(subset_filtered)
```
